## Etapa 2 — Limpeza e Tratamento dos Dados (Python / Pandas)

```python
#Tratar valores nulos (substituir ou remover conforme o contexto)

df['desconto'] = df['desconto'].fillna(0)

print('Valores Nulos após tratamento:', df.isnull().sum())
```

```python
#Notas Nulas o ciente pode apenas não ter avaliado

df['avaliacao_cliente'] = pd.to_numeric(df['avaliacao_cliente'], errors='coerce')

nao_avaliados = df['avaliacao_cliente'].isnull().sum()
print(f'Clientes que não avaliaram: {nao_avaliados}')
```

```python
#Converter a coluna data_venda para o tipo datetime

df['data_venda'] = pd.to_datetime(df['data_venda'])

df
```

```python
#Extrair novas colunas: mes, ano, trimestre a partir de data_venda

#mês

df['mes'] = df['data_venda'].dt.month_name(locale='pt_BR')

df.head()
```

```python
#ano

df['ano'] = df['data_venda'].dt.year

df.head()
```

```python
#trimestre

df['trimestre'] = df['data_venda'].dt.quarter

df.head()
```

```python
#Verificar e remover duplicatas, se existirem

df.duplicated().sum()
```

```python
#Padronizar textos, descobrir quais colunas são de texto.

colunas_texto = df.select_dtypes(include='object').columns

print('Colunas de Texto:', colunas_texto)
```

```python
##Padronizar textos (ex: remover espaços, corrigir capitalização com .str.strip() e .str.title())

for coluna in colunas_texto:
    df[coluna] = df[coluna].str.strip().str.title()
```

```python
#Validar se valor_total está coerente com quantidade * valor_unitario * (1 - desconto)

df['valor_total'] = (df['quantidade'] * df['valor_unitario'] * (1 - df['desconto'])).round(2)

df.head()
```

## Etapa 3 — Análise Exploratória de Dados — EDA (Python / Pandas / NumPy)

```python
## Análise Geral
#1 Qual foi o faturamento total no período analisado?

faturamento_total = df['valor_total'].sum()

print(f'Faturamento Total: {faturamento_total:,.2f}')
```

```python
#Qual foi o ticket médio por venda?

ticket_medio = df['valor_total'].mean()

print(f'Ticket Médio por Venda: {ticket_medio:,.2f}')
```

```python
#Quantas vendas foram realizadas no total?

total_vendas = df.shape[0]

print(f'Total de Vendas: {total_vendas}')
```

```python
## Análise Temporal

#4Como o faturamento evoluiu mês a mês? Houve sazonalidade?

faturamento_mensal = df.groupby(['ano', 'mes'])['valor_total'].sum().reset_index()

faturamento_mensal
```

```python
#5 Qual foi o mês com maior e menor volume de vendas?

vendas_mensais = df.groupby(['ano', 'mes'])['quantidade'].sum().reset_index()

mes_maior_vendas = vendas_mensais.loc[vendas_mensais['quantidade'].idxmax()]
mes_menor_vendas = vendas_mensais.loc[vendas_mensais['quantidade'].idxmin()]

print(f'Mês com maior volume de vendas: {mes_maior_vendas["mes"]} ({mes_maior_vendas["quantidade"]})')

print(f'Mês com menor volume de vendas: {mes_menor_vendas["mes"]} ({mes_menor_vendas["quantidade"]})')
```

```python
##Análise por Categoria e Produto
#6 Qual categoria gerou maior receita?

receita_categoria = df.groupby('categoria')['valor_total'].sum().reset_index()
categoria_maior_receita = receita_categoria.loc[receita_categoria['valor_total'].idxmax()]

print(f'Categoria com maior receita: {categoria_maior_receita["categoria"]} ({categoria_maior_receita["valor_total"]:,.2f})')
```

```python
#7 Quais são os 5 produtos mais vendidos em quantidade?

produtos_mais_vendidos = df.groupby('produto')['quantidade'].sum().reset_index().sort_values(by='quantidade', ascending=False).head(5)

print('5 Produtos mais vendidos em quantidade:')
print(produtos_mais_vendidos)
```

```python
#8 Qual categoria tem a maior média de avaliação dos clientes?

categoria_maior_media_avaliacao = df.groupby('categoria')['avaliacao_cliente'].mean().reset_index().sort_values(by='avaliacao_cliente', ascending=False).head(1)

print('Categoria com a maior média de avaliação dos clientes:')
print(categoria_maior_media_avaliacao)  
```

```python
#Análise Geográfica
# 9 Quais são os 5 estados com maior faturamento?

faturamento_estados = df.groupby('estado')['valor_total'].sum().reset_index().sort_values(by='valor_total', ascending=False).head(5)

faturamento_estados
```

```python
#10 Existe relação entre a cidade/estado e a forma de pagamento preferida?

preferencia_pagamento = (
    df.groupby(['estado', 'forma_pagamento'])
    .size()
    .reset_index(name='quantidade_vendas')
    .sort_values(['estado', 'quantidade_vendas'], ascending=[True, False])

)

preferencia_pagamento.head(20)
```

```python
## Análise de Clientes
#11 Qual a distribuição das formas de pagamento utilizadas?

df.groupby('forma_pagamento').size()
```

```python
#12 Qual a média de avaliação geral dos clientes? Como ela varia por categoria?

media_avaliacao_geral = df['avaliacao_cliente'].mean()

print(f'Média geral de avaliação: {media_avaliacao_geral:.2f}')
```

```python
# Como ela varia por categoria?

media_categoria_avaliacao = df.groupby('categoria')['avaliacao_cliente'].mean().reset_index().sort_values(by='avaliacao_cliente', ascending=False)

media_categoria_avaliacao
```
